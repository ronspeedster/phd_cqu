```python
import subprocess, os
from pathlib import Path
import subprocess, shutil

fasta = Path("data/processed_grouped/bins_aligned/group_002_aligned.fasta").resolve()
jalview = shutil.which("jalview")


def run_command_silently(command):
    with open(os.devnull, 'w') as devnull:
        subprocess.run(command, stdout=devnull, stderr=devnull)


if jalview is None:
    print("Jalview executable not found on PATH. Install Jalview Desktop or add it to PATH.")
else:
    # subprocess.Popen([jalview, str(fasta)])
    run_command_silently([jalview, str(fasta)])
```
